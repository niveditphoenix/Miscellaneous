# Groupby output format â€“ Series or DataFrame?

```python
df_phone.groupby('month')['duration'].sum() # produces Pandas Series
```

```python
df_phone.groupby('month')[['duration']].sum() # produces Pandas DataFrame
```

The groupby output will have an index or multi-index on rows corresponding to your chosen grouping variables. 

To avoid setting this index, pass **â€œas_index=Falseâ€** to the groupby operation.

```python
df_phone.groupby('month', as_index=False)['duration'].sum()
```

# Multiple Statistics Per Group using -> 'agg()' function
It is similar to that of MongoDBâ€™s aggregation framework.

```python
# Group the data frame by month and item and extract a number of stats from each group
df_phone.groupby(['month','item']).agg({'duration':'sum', 
                                        'network_type':'count', 
                                        'date':'first'})
```

```python
# Same as above with last date
df_phone.groupby(['month','item']).agg({'duration':'sum', 
                                        'network_type':'count', 
                                        'date':'last'})
```

```python
# Define the aggregation procedure outside of the groupby operation
aggregations = {'duration':'sum',
                'date': 'first'}

df_phone.groupby('month').agg(aggregations)
```

## Applying multiple functions to columns in groups

```python
# Group the data frame by month and item and extract a number of stats from each group
aggregations = {'duration':[min, max, sum, 'mean', 'median'],
                'network_type':['count', 'nunique'], 
                'date':[min, 'first', 'last', 'nunique']}

df_phone.groupby(['month', 'item']).agg(aggregations)
```

### Renaming grouped statistics from groupby operations

- When multiple statistics are calculated on columns, the resulting dataframe will have a multi-index set on the column axis. 
- This can be difficult to work with.

#### A neater approach, is using 
- The **ravel()** method on the grouped columns.
- **Ravel()** turns a Pandas multi-index into a simpler array, which we can combine into sensible column names

```python
aggregation = {'duration':[min, max,'mean'], 
               'date':['first', 'last']}

grouped = df_phone.groupby('month').agg(aggregation)        # agg function will carry the dictionary always

# Using ravel, and a string join, we can create better names for the columns:
grouped.columns = ['_'.join(x) for x in grouped.columns.ravel()]

grouped.head()
```

Below syntax will be deprecated in newer pandas version.

In older Pandas releases (< 0.20.1), **rename the newly calculated columns** was possible through nested dictionaries, or by passing a list of functions for a column. 

Our final example calculates multiple values from the duration column and names the results appropriately. 

```python
# Define the aggregation calculations
aggregations = {
    'duration': { # work on the "duration" column
        'total_duration': 'sum',  # get the sum, and call this result 'total_duration'
        'average_duration': 'mean', # get mean, call result 'average_duration'
        'num_calls': 'count'
    },
    'date': {     # Now work on the "date" column
        'max_date': 'max',   # Find the max, call the result "max_date"
        'min_date': 'min',
        'num_days': lambda x: max(x) - min(x)  # Calculate the date range per group
    }
}

df_phone[df_phone['item'] == 'call'].groupby('month').agg(aggregations)
```
