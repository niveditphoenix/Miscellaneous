```python
# Now let's split train and test
trainX, testX, trainY, testY = train_test_split(X_std_df.values, Y, test_size=0.2, random_state=123)
```

```python
for i in [trainX, testX, trainY, testY, X_std_df, Y]:
    i.shape
```

```python
trainX[:4]
trainY[:4]
```

```python
from sklearn.metrics import confusion_matrix, mean_squared_error
```

```python
# Function to get diffenet model and fit it
def fun_alg(clf):
    model = clf.fit(trainX, trainY)
    return model
```

```python
# Funtion to predict on train and test basis passed model
def fun_predict(model):
    pred_train = model.predict(trainX) 
    pred_test = model.predict(testX)
    return pred_train, pred_test
```

```python
# Function to get accuracy for train and test basis predicted resukts and return these
def fun_RMSE(model):
    pred_train, pred_test = fun_predict(model)
    accuracy_train = np.round(np.sqrt(mean_squared_error(trainY, pred_train)),4)
    accuracy_test = np.round(np.sqrt(mean_squared_error(testY, pred_test)),4)
    return accuracy_train, accuracy_test
```

## Checking for 3 algos:
- Random Forest
- Gradient Boosting
- xgBoost

```python
clf_RF = RandomForestRegressor(n_estimators=10, max_depth=10, n_jobs=4, random_state=123, verbose=2)
clf_GB = GradientBoostingRegressor(learning_rate=0.1, n_estimators=10, max_depth=10, random_state=123)
clf_xgb = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, 
                           max_depth=3, learning_rate=0.01, n_estimators=100, alpha=10, random_state=123)
```

```python
# Gettiong each model
for i in [clf_RF, clf_GB, clf_xgb]:
    print('*****   MODEL   *****')
    model = fun_alg(i)
    pred_train, pred_test = fun_predict(model)
    accuracy_train, accuracy_test = fun_RMSE(model)
    #print(pred_train, pred_test)
    print('Train RMSE is {0} & \n Test RMSE is {1}'.format(accuracy_train, accuracy_test))
    print('********   END OF MODEL   *******')
    print("\n")
```

Currently basis above, we are getting MAX RMSE for xgboost and least for Random Forest and that too when we used just 10 estimators for RF while 100 for xgBoost

# k-fold Cross Validation using XGBoost

#### Converting the dataset into an optimized data structure called Dmatrix that XGBoost supports and gives it acclaimed performance and efficiency gains.

```python
data_dmatrix = xgb.DMatrix(data=X_std_df, label=Y)
```

```python
# Creating Hyper-Parameter Dictionary 
param = {"objective":"reg:linear", 'colsample_bytree': 0.3, "learning_rate":0.1, "alpha":10, "max_depth":5}
```

```python
cv_results = xgb.cv(dtrain=data_dmatrix, params=param, nfold=10, 
                    num_boost_round=50, metrics='rmse', as_pandas=True, seed=123)
```

```python
cv_results.head()
```

```python
cv_results.tail()
```

```python
cv_results['test-rmse-mean'].tail()
```

Thus, with CV, we are getting RMSE for last record i.e. 49th to be 3.6 which is even lesser than RF.

Can explroe more with Grid Search

# Visualize Boosting Trees and Feature Importance

```python
xg_reg = xgb.train(params=param, dtrain=data_dmatrix, num_boost_round=10)
```

Visualizing the importance of each features in this dataset

```python
plt.rcParams['figure.figsize'] = [20, 5]
xgb.plot_importance(xg_reg)
plt.show()
```

Can see that **RM - average number of rooms per dwelling** has been given the MOST IMPORTANCE among all features.

- RM    -> average number of rooms per dwelling
- NOX   -> nitric oxides concentration (parts per 10 million)
- DIS   -> weighted distances to five Boston employment centres
- LSTAT -> % lower status of the population
- TAX   -> full-value property-tax rate per $10,000
